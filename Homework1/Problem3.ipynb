```python
under = 1
over = 1
for i in range(100000):
    under= float(under/2)
    over= float(over*2)
    print(i, under, over)
```

```python
import numpy as np

under = 1
over = 1

for i in range(100000):
    under= np.float32(under/2)
    over= np.float32(over*2)
    print(i, under, over)
```

```python
under = -np.int32(2**31 - 5)
over = np.int32(2**31 - 5)
i=1
while (i<10):
    print("under: ",under," over: ", over, " i: ", i)
    under= np.int32(under-1)
    over= np.int32(over+1)
    i+=1
    
```

```python
under = -np.int64(2**63 - 5)
over = np.int64(2**63 - 5)
i=1
while (i<10):
    print("under: ",under," over: ", over, " i: ", i)
    under= np.int64(under-1)
    over= np.int64(over+1)
    i+=1
    
```
